```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import numpy as np
```

```python
df = pd.read_csv("C:\\Users\\devik\\Prodigy\\Task_4\\Tweets.csv")
df
```

```python
df['tweet_created'] = pd.to_datetime(df['tweet_created'], format='%d-%m-%Y %H:%M')
```

```python
# Sentiment Distribution
sentiment_counts = df['airline_sentiment'].value_counts()
plt.figure(figsize=(8, 6))
plt.pie(sentiment_counts, labels=sentiment_counts.index, autopct='%1.1f%%', startangle=140, colors=['grey', 'green', 'red'])
plt.title('Sentiment Distribution')
plt.show()
```

```python
# Sentiment counts
sentiment_counts = df['airline_sentiment'].value_counts()

# Plot sentiment distribution
plt.figure(figsize=(10, 6))
sns.barplot(x=sentiment_counts.index, y=sentiment_counts.values, palette='viridis')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment')
plt.ylabel('Number of Tweets')
plt.show()

```
