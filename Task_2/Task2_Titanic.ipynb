```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Titanic dataset
df = pd.read_csv('C:\\Users\\devik\\Prodigy\\Titanic-Dataset.csv')
print(df.head())
```

```python
#Data Cleaning
# Check for missing values
print(df.isnull().sum())

# Visualize missing values
plt.figure(figsize=(8, 6))
sns.heatmap(df.isnull(), cmap='viridis', yticklabels=False, cbar=False)
plt.title('Missing Data: Titanic Dataset')
plt.show()

# Handling missing values in 'Age' column (filling with median age)
df['Age'].fillna(df['Age'].median(), inplace=True)

# Handling missing values in 'Embarked' column (filling with mode)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Dropping 'Cabin' column due to too many missing values
df.drop('Cabin', axis=1, inplace=True)

# Verify if missing values are handled
print(df.isnull().sum())
```

```python
#Distribution of Survivors
# Countplot of survivors
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', data=df)
plt.title('Survival Count (0 = No, 1 = Yes)')
plt.show()
```

```python
#Survival by Gender
# Countplot of survivors by gender
plt.figure(figsize=(6, 4))
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title('Survival Count by Gender')
plt.legend(title='Sex', loc='best')
plt.show()
```

```python
#Age Distribution of Passengers
# Distribution plot of passenger ages
plt.figure(figsize=(8, 6))
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

```

```python
#Correlation Heatmap
# Select numeric columns for correlation computation
numeric_cols = df.select_dtypes(include=np.number)

# Correlation heatmap of numeric variables
plt.figure(figsize=(8, 6))
sns.heatmap(numeric_cols.corr(), annot=True, cmap='coolwarm', linewidths=.5)
plt.title('Correlation Heatmap of Numeric Variables')
plt.show()
```
