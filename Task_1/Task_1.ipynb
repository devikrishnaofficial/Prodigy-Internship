```python
import pandas as pd

# Load the Titanic dataset
df = pd.read_csv('C:\\Users\\devik\\Prodigy\\Task_1\\Titanic-Dataset.csv')
print(df.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the visual aesthetic
sns.set(style="whitegrid")

# Create a histogram for the Age column
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'].dropna(), bins=30, kde=False, color='blue')

# Add titles and labels
plt.title('Distribution of Ages on the Titanic')
plt.xlabel('Age')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
# Create a bar chart for the Gender column
plt.figure(figsize=(8, 5))
sns.countplot(x='Sex', data=df, palette='Set2')

# Add titles and labels
plt.title('Distribution of Genders on the Titanic')
plt.xlabel('Gender')
plt.ylabel('Count')

# Show the plot
plt.show()
```
